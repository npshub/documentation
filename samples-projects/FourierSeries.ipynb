## Fourier Series

Describe Fourier Series here.....

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

### Even Function

Describe even function here....

```python
R = 100
```

```python
DFC = pd.DataFrame({"cosa":[1*np.cos(1*np.pi/18*i) for i in range(R)],\
                   "cos2a":[1/2*np.cos(2*np.pi/18*i) for i in range(R)],\
                   "cos3a":[1/4*np.cos(3*np.pi/18*i) for i in range(R)],\
                   "cos4a":[1/8*np.cos(4*np.pi/18*i) for i in range(R)]})
```

```python
DFC.head()
```

```python
DFC.plot(figsize = [10,4])
```

### Odd Function

Describe odd function here....

```python
DFS = pd.DataFrame({"sina":[1*np.sin(1*np.pi/18*i) for i in range(R)],\
                   "1/2 sin2a":[1/2*np.sin(2*np.pi/18*i) for i in range(R)],\
                   "1/4 sin3a":[1/4*np.sin(3*np.pi/18*i) for i in range(R)],\
                   "1/8 sin4a":[1/8*np.sin(4*np.pi/18*i) for i in range(R)]})
```

```python
DFS.head()
```

```python
DFS.plot(figsize = [10,4])
```
