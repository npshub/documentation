### Plot of different waves motion

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import glob
import moviepy.editor as mpy
from mpl_toolkits.mplot3d import axes3d
import seaborn as sns
sns.set()
```

- Define Amplitude, angular frequency, wave vector

```python
A = 1.0
w = 2*np.pi
k = 2*np.pi
```

- Calcualte frequency, weblength and velocity

```python
f = w/(2*np.pi)
wl = 2*np.pi/k
v = f*wl
(f,wl,v)
```

#### Plot the different configuration of wave motion

```python
x = np.arange(0,5,0.05)
t = np.arange(0,5,0.05)
```

#### Plot the particle displacement keeping time fixed
When we fix the time it is like taking a photograph of the wave motion. It can be represent as

$Y = A \sin (kx + \phi_{0})$

A special case is **closed vibrating string**

```python
y = A*np.sin(w*1.0 - k*x)
```

```python
plt.figure(figsize = [12,4])
plt.stem(y,use_line_collection=True)
plt.xlabel("Distance")
plt.ylabel("Amplitude")
plt.title('Particle displacements at t=2.0')
```

```python
Data = []
for xx in x:
    Data.append({"wave1": A*np.sin(w*1.0 - k*xx),\
                "wave2": A*np.sin(w*1.25 - k*xx),\
                "wave3": A*np.sin(w*1.50 - k*xx),\
                "wave4": A*np.sin(w*1.75 - k*xx),\
                "wave5": A*np.sin(w*2.0 - k*xx)})
```

```python
DF = pd.DataFrame(Data)
DF.head()
```

```python
DF.plot(figsize = [12,4])
```

#### Plot the particle displacement keeping position fixed

When we fix the position and look at a single particle displacement, it behaves like a harmonic oscillator. It can be represent as

$Y = A \sin (\omega t + \phi_{0})$

A special case is **harmonic oscillator**

```python
y = A*np.sin(w*t - k*1.0)
```

```python
plt.figure(figsize = [12,4])
plt.stem(y,use_line_collection=True)
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title('Particle motion at x=2.0')
```
